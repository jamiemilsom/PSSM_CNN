```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.read_csv('mega2.txt',delimiter='\t')
results_b = pd.read_csv('mega2b.txt',delimiter='\t')
temp1 = results.drop(columns=['Trial name','loss'])
temp2 = results_b.drop(columns=['Trial name', 'status', 'loc', 'batch_size','epochs','test accuracy','total time (s)', 'iter','loss'])
master_results = pd.concat([temp1,temp2],axis=1)

threshold_difference = 0.005
mask = master_results['test accuracy'] < (master_results['val accuracy'] + threshold_difference)
filtered_results = master_results[mask]
filtered_results.sort_values('val accuracy')
print(filtered_results['val accuracy'].max())
max_val_accuracy_index = filtered_results['val accuracy'].idxmax()
print(filtered_results.loc[max_val_accuracy_index])

plt.scatter(filtered_results.index,filtered_results['val accuracy'])
plt.scatter(filtered_results.index,filtered_results['test accuracy'])
plt.ylim(0.76,0.81)
```
